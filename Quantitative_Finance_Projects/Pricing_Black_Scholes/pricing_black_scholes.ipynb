```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})
```

```python
def black_scholes_call_price(S,K,T,r,sigma) :
    d1 = (np.log(S/K) + (r+(sigma**2)/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N_d1 = norm.cdf(d1)
    N_d2 = norm.cdf(d2)
    return S*N_d1 - K*np.exp(-r*T)*N_d2
```

```python
sigmas = np.linspace(0.01,1,400)
prices = [black_scholes_call_price(S=100,K=120,T=2,r=0.05,sigma=s) for s in sigmas]
plt.plot(sigmas,prices)
plt.xlabel(r"\textbf{Volatility} ($\sigma$)")
plt.ylabel(r"\textbf{Call price}")
plt.title(r"\textbf{European Call price vs Volatility (Black-Scholes)}")
plt.grid(True)
plt.show()
```

```python
exercise_dates = np.linspace(0.01,5,100)
sigmas=[0.3,0.7,1,1.3,1.6]
for s in sigmas :
    plt.plot(exercise_dates,[black_scholes_call_price(S=100,K=120,T=t,r=0.05,sigma=s) for t in exercise_dates], label = rf"\textbf{{Volatility}} ($\sigma$) ={s}")

plt.xlabel(r"\textbf{exercise dates (T)}")
plt.ylabel(r"\textbf{Call price}")
plt.legend()
plt.title(r"\textbf{European Call price vs exercise dates (Black-Scholes)}")
plt.grid(True)
plt.show()
```
